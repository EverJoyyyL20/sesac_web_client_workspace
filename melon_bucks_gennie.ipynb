```python
from urllib.request import urlopen
import bs4

url="https://music.bugs.co.kr/chart"
html=urlopen(url)
html
```

```python
bs_obj=bs4.BeautifulSoup(html,'html.parser')
type(bs_obj.prettify()[:2468])
```

```python
print(bs_obj.prettify()[:2468])
```

### bucks_ranking

```python
ranking_num = []
ranks = bs_obj.select('.ranking strong')   # <-- 리스트(요소들)

for rank in ranks:
    ranking_num.append(rank.get_text(strip=True))   # 각 요소에서 text 추출

print(ranking_num)



```

### bucks_titles

```python
title_str= []
titles = bs_obj.select('.title a')   # <-- 리스트(요소들)

for title in titles[2:]:
    title_str.append(title.text)

```

```python
title_str
```

### bucks_artist

```python
artist_str=[]
artists = bs_obj.select('.artist a')

for a in artists:
    artist_str.append(a['title'])
```

```python
artist_str  = [a for a in artist_str if a != '아티스트 전체보기']
```

```python
import pandas as pd
df_bucks=pd.DataFrame({"순위":ranking_num,"구분":"벅스",'타이틀':title_str,'가수':artist_str},)
df_bucks
```

## melon

```python
import requests
import bs4

url = "https://www.melon.com/chart/index.htm"


headers = {
    'User-Agent': "User-Agent",
    "Referer": "https://www.melon.com/"
}

response = requests.get(url, headers=headers)

bs_obj = bs4.BeautifulSoup(response.text, "html.parser")

print(bs_obj.prettify()[:500])
```

### melon ranking

```python
rank_num=[]
ranks=bs_obj.find_all("span",{'class':'rank'})

for rank in ranks[1:]:
    rank_num.append(rank.text)

rank_num
```

### melon song title

```python
m_song_title=[]
songs=bs_obj.find_all("div",{"class":"ellipsis rank01"})


for song in songs:
    a_tag=song.find('a')
    m_song_title.append(a_tag['title'].strip('재생'))

print(m_song_title)

```

### melon singer

```python
m_singer=[]

singers=bs_obj.find_all("div",{"class":"ellipsis rank02"})

for singer in singers:
    singer=singer.find('a')
    m_singer.append(singer["title"].strip("- 페이지 이동"))
```

```python
df_melon=pd.DataFrame({"순위":rank_num,"구분":"멜론","타이틀":m_song_title,"가수":m_singer})

df_melon
```

```python
df_melon
```

```python
df_melon.to_excel("melon.xlsx")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
rank_num=[]
title_str=[]
singer_str=[]
rank_num2=[]
def genie_data(page):
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    driver.get("https://www.genie.co.kr/chart/top200?ditc=D&ymd=20251209&hh=16&rtm=Y&pg="+str(page))
    genie_datas=driver.find_elements(By.CSS_SELECTOR,'.list-wrap tr.list')

    for genie_data in genie_datas:
        rank=genie_data.find_elements(By.CSS_SELECTOR,'td.number')
        rank_num.append(rank[0].text[:3].split("\n"))
        title=genie_data.find_elements(By.CSS_SELECTOR,'a.title.ellipsis')
        title_str.append(title[0].text)
        singer=genie_data.find_elements(By.CSS_SELECTOR,'a.artist.ellipsis')
        singer_str.append(singer[0].text)





```

```python
genie_data(1)
```

```python
genie_data(2)
```

```python
rank_num2=[]
for i in rank_num:
    rank_num2.append(i[0])

```

```python
rank_num2
```

```python
singer_str
```

```python
import pandas as pd
gennie_df=pd.DataFrame({"순위":rank_num2,"구분":"지니","타이틀":title_str,"가수":singer_str})
```
