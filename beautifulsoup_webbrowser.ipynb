```python
def replace_newline(soup_html):
    br_to_newlines=soup_html.find_all("br") # bríƒœê·¸ ëª¨ë‘ ì°¾ê¸°
    for br_to_newline in br_to_newlines:
        br_to_newline.replace_with("\n") # ì°¾ì€ bríƒœê·¸ë¥¼ ëª¨ë‘ \nìœ¼ë¡œ ë°”ê¿ˆ ë°˜ë³µë¬¸ìœ¼ë¡œ 
    return soup_html
```

```python
import requests
from bs4 import BeautifulSoup
import os

def get_image_url(url):
    html_image_url=requests.get(url).text
    soup_image_url=BeautifulSoup(html_image_url,"lxml")
    image_elements=soup_image_url.select('img')
    if (image_elements !=None):
        image_urls=[]
        for image_element in image_elements:
            image_urls.append(image_element.get('src'))
        return image_urls
    else:
        return None
    
```

```python
def download_image(img_folder,img_url):
    if(img_url != None):
        html_image = requests.get(img_url)   # stream=True ì¶”ê°€
        imageFile = open(os.path.join(img_folder, os.path.basename(img_url)), 'wb')
        chunk_size = 1000000

        for chunk in html_image.iter_content(chunk_size):   # iter.content â†’ iter_content ìˆ˜ì •
            if chunk:
                imageFile.write(chunk)

        imageFile.close()   # close()ëŠ” forë¬¸ ë°–ì—ì„œ ì‹¤í–‰í•´ì•¼ í•¨

        print("ì´ë¯¸ì§€ íŒŒì¼ëª… :'{0}', ë‚´ë ¤ë°›ê¸° ì™„ë£Œ!".format(os.path.basename(img_url)))
    else:
        print("ë‚´ë ¤ë°›ì„ ì´ë¯¸ì§€ê°€ ì—†ìŠµë‹ˆë‹¤.")

reshot_url='https://www.reshot.com/search/animal'

figure_folder="C:/Myexam/download"

reshot_image_urls=get_image_url(reshot_url)

num_of_download_image=3

for k in range(num_of_download_image):
    download_image(figure_folder,reshot_image_urls[k])
print("=============")
print("ì„ íƒí•œ ëª¨ë“  ì´ë¯¸ì§€ ë‚´ë ¤ë°›ê¸° ì™„ë£Œ!")
```
